```python
df=pd.get_dummies(df,columns=['State'])
```

```python
x=df.drop(['Profit'],axis=1)
y=df['Profit']
```

```python
x.head()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y)
```

```python
reg=LinearRegression()
reg.fit(x_train,y_train)
```

```python
y_pred=reg.predict(x_test)
```
